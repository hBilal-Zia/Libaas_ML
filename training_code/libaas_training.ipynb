```python
import os
import numpy as np
import pandas as pd
import glob
from keras.applications import ResNet50
from keras.layers import GlobalAveragePooling2D, Dense, Dropout
from keras.models import Model
from keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
print(os.curdir)
```

```python
test_path = '/content/test/test'
train_path = '/content/train/train'
val_path = '/content/val/val'

wear_types = ['bottomwear', 'upperwear', 'footwear']
```

```python
def directory_to_csv(base_dir):
    data = []
    wear_types = ['bottomwear', 'upperwear', 'footwear']

    # Loop through each category
    for wear_type in wear_types:
        category_path = os.path.join(base_dir, wear_type)

        # Get all subcategories (subfolders) within the category
        article_types = os.listdir(category_path)

        for article_type in article_types:
            article_type_path = os.path.join(category_path, article_type)

            # Get all image files in the subcategory folder
            image_files = glob.glob(os.path.join(article_type_path, '*.*'))  # Matches all file types

            for image_file in image_files:
                # Add the image file and its labels (category and subcategory) to the list
                data.append({
                    'image': image_file,
                    'wearType': wear_type,
                    'articleType': article_type
                })

    # Create a DataFrame from the data
    df = pd.DataFrame(data)

    return df
```

```python
df_test = directory_to_csv(test_path)
df_train = directory_to_csv(train_path)
df_val = directory_to_csv(val_path)
df_test.to_csv('items_test.csv')
df_train.to_csv('items_train.csv')
df_val.to_csv('items_val.csv')
```

```python
print(df_test.shape)
print(df_train.shape)
print(df_val.shape)
```

```python
img_hieght, img_width = 224, 224
batch_size = 512
```

```python
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(img_hieght, img_width, 3))
for layer in base_model.layers:
    layer.trainable = False
```

```python
datagen = ImageDataGenerator(
            rescale=1./255,
            rotation_range=20,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='nearest'
)

    # Create a data generator
generator_train = datagen.flow_from_dataframe(
        df_train,
        x_col='image',
        y_col='articleType',
        target_size=(img_hieght, img_width),
        batch_size=batch_size,
        class_mode='categorical'
)
```

```python
datagen = ImageDataGenerator(
            rescale=1./255
)

    # Create a data generator
generator_val = datagen.flow_from_dataframe(
        df_val,
        x_col='image',
        y_col='articleType',
        target_size=(img_hieght, img_width),
        batch_size=batch_size,
        class_mode='categorical'
)
```

```python
print(generator_train.class_indices)
class_names = generator_train.class_indices
labels = dict((v, k) for k, v in class_names.items())
print(class_names)
print(type(labels))
print(labels)
print(list(labels.values()))
```

```python

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
predictions = Dense(10, activation='softmax')(x)
```

```python
model = Model(inputs=base_model.input, outputs=predictions)

# Compile the model
model.compile(optimizer=Adam(learning_rate=1e-4), loss='categorical_crossentropy', metrics=['accuracy'])

# Summary of the model architecture
model.summary()

```

```python
history =  model.fit(generator_train, validation_data=generator_val, epochs=1, verbose=1)

```
